```python
from dataclasses import dataclass

@dataclass
class FieldStats:
    minimum: float = 0.0
    maximum: float = 0.0
    n_missing: int = 0

def get_stats(series: pd.Series) -> FieldStats:
    return FieldStats(
        minimum = series.min(),
        maximum = series.max(),
        n_missing = series.isna().value_counts().get(True, 0)
    )
```

```python
get_stats(df['objra'])
```

```python
%%time
def get_pq_stats(pq_file: Path, field_name: str) -> FieldStats:
    df = pd.read_parquet(pq_file, columns=[field_name], engine='pyarrow')
    return get_stats(df[field_name])

# Having to do this instead of a lambda because multiprocessing cannot
# pickle lambdas when using separate workers
def shim(pq_file: Path):
    return get_pq_stats(pq_file, 'objra')


import multiprocessing
with multiprocessing.Pool(32) as pool:
    stats = pool.map(shim, pqs)
stats
```
